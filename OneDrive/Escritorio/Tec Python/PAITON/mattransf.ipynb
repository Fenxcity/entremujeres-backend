```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A1=np.array([[0.5,0],[0,2]])
M=np.array([[0,0.5,1,0],[0,1,0,0]])

plt.plot(M[0],M[1],'-o')
plt.axis('equal')
plt.show()
```

```python
A2=np.array([[1,0],[0,-1]])
W=np.matmul(A2,np.matmul(A1,M))
plt.plot(W[0],W[1],'-o')
plt.axis('equal')
plt.show()
```

```python
theta=np.pi/4
A3=np.array([[np.cos(theta),-np.sin(theta)],[np.sin(theta),np.cos(theta)]])
W=np.matmul(A3,np.matmul(A2,np.matmul(A1,M)))
plt.plot(W[0],W[1],'-o')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ============================================================
#  PASO 1: DEFINIR LA FIGURA DEL POLLO COMO MATRICES (2×N)
#  Cada parte es un arreglo 2D:  fila 0 = x,  fila 1 = y
# ============================================================

t = np.linspace(0, 2 * np.pi, 80)          # parámetro para curvas

# --- Cuerpo (elipse grande) ---
cuerpo = np.array([
    2.0 * np.cos(t),                        # x
    1.3 * np.sin(t) + 1.5                   # y  (centrada en y=1.5)
])

# --- Cabeza (círculo) ---
cabeza = np.array([
    0.85 * np.cos(t) + 2.3,                 # x  (desplazada a la derecha)
    0.85 * np.sin(t) + 3.0                  # y  (arriba del cuerpo)
])

# --- Pico (triángulo) ---
pico = np.array([
    [3.0, 3.7, 3.0, 3.0],                  # x
    [3.2, 3.0, 2.8, 3.2]                   # y
])

# --- Cresta (zigzag rojo) ---
cresta = np.array([
    [1.8, 2.0, 2.1, 2.3, 2.4, 2.6, 2.8],  # x
    [3.7, 4.2, 3.9, 4.3, 4.0, 4.2, 3.7]   # y
])

# --- Barbilla (debajo del pico) ---
barbilla = np.array([
    0.18 * np.cos(t) + 2.9,
    0.22 * np.sin(t) + 2.5
])

# --- Ojo (círculo pequeño) ---
ojo = np.array([
    0.12 * np.cos(t) + 2.45,
    0.12 * np.sin(t) + 3.2
])

# --- Ala (media elipse interior) ---
t_ala = np.linspace(-np.pi / 2, np.pi / 2, 50)
ala = np.array([
    1.1 * np.cos(t_ala) + 0.2,
    0.55 * np.sin(t_ala) + 1.5
])

# --- Cola (plumas) ---
cola1 = np.array([[-1.9, -2.8, -3.2, -2.4], [1.8, 2.8, 3.8, 2.6]])
cola2 = np.array([[-1.9, -3.0, -3.5, -2.6], [1.5, 2.3, 3.3, 2.2]])
cola3 = np.array([[-1.9, -2.6, -3.0, -2.2], [2.1, 3.2, 4.1, 3.0]])

# --- Patas (líneas + dedos) ---
pata_izq = np.array([
    [-0.4, -0.5, -0.9, -0.5, -0.1, -0.5],  # x
    [ 0.2, -0.6, -0.9, -0.6, -0.9, -0.6]   # y
])
pata_der = np.array([
    [0.6, 0.5, 0.1, 0.5, 0.9, 0.5],        # x
    [0.2, -0.6, -0.9, -0.6, -0.9, -0.6]     # y
])


# ============================================================
#  PASO 2: GRAFICAR LA FIGURA ORIGINAL
# ============================================================

fig, ax = plt.subplots(figsize=(8, 8))

# Cuerpo y cabeza rellenos
ax.fill(cuerpo[0], cuerpo[1], color='#FFD966', ec='#B8860B', lw=2)
ax.fill(cabeza[0], cabeza[1], color='#FFD966', ec='#B8860B', lw=2)

# Cresta y barbilla
ax.fill(cresta[0], cresta[1], color='#E74C3C', ec='#C0392B', lw=2)
ax.fill(barbilla[0], barbilla[1], color='#E74C3C', ec='#C0392B', lw=1.5)

# Pico
ax.fill(pico[0], pico[1], color='#F39C12', ec='#D68910', lw=2)

# Ojo (blanco + pupila negra)
ax.fill(ojo[0], ojo[1], color='white', ec='black', lw=1.5)
ax.plot(2.45, 3.2, 'ko', ms=4)

# Ala
ax.fill(ala[0], ala[1], color='#F0C040', ec='#B8860B', lw=1.5)

# Plumas de la cola
for cola_i in [cola1, cola2, cola3]:
    ax.fill(cola_i[0], cola_i[1], color='#E67E22', ec='#A04000', lw=2)

# Patas
ax.plot(pata_izq[0], pata_izq[1], color='#D4840A', lw=3, solid_capstyle='round')
ax.plot(pata_der[0], pata_der[1], color='#D4840A', lw=3, solid_capstyle='round')

ax.set_title('🐔  Pollo original  (Paso 1: Definición matricial)', fontsize=14)
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.set_xlim(-5, 5.5)
ax.set_ylim(-2, 5.5)

plt.tight_layout()
plt.savefig('/mnt/user-data/outputs/pollo_original.png', dpi=150)
plt.show()
print("✅ Figura guardada como pollo_original.png")
```
